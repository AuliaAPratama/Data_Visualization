### Harga Historis Saham Provider Telco Indonesia

```python
# tanpa indexing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dfXl = pd.read_csv('EXCL.JK.csv',parse_dates=['Date'])
dfXl = dfXl.replace(0,np.NaN)
dfXl = dfXl.fillna(method='ffill')

dfFren = pd.read_csv('FREN.JK.csv',parse_dates=['Date'])
dfFren = dfFren.replace(0,np.NaN)
dfFren = dfFren.fillna(method='ffill')

dfIsat = pd.read_csv('ISAT.JK.csv',parse_dates=['Date'])
dfIsat = dfIsat.replace(0,np.NaN)
dfIsat = dfIsat.fillna(method='ffill')

dfTlkm = pd.read_csv('TLKM.JK.csv',parse_dates=['Date'])
dfTlkm = dfTlkm.replace(0,np.NaN)
dfTlkm = dfTlkm.fillna(method='ffill')

tanggal1 = pd.date_range(start='2019-03-16', end='2019-06-15', freq='W-sat')

plt.figure(figsize=(12,5), dpi = 75)
plt.plot(dfXl['Date'],dfXl['Close'], color = 'green', label = 'PT XL AXIATA TBK' )
plt.plot(dfFren['Date'],dfFren['Close'], color = 'lightblue', label = 'PT Smartfren Telecom TBK')
plt.plot(dfIsat['Date'],dfIsat['Close'], color = 'blue', label = 'PT Indosat TBK')
plt.plot(dfTlkm['Date'],dfTlkm['Close'], color = 'red', label = 'PT Telekomunikasi Indonesia TBK')

plt.suptitle("Harga Historis Saham Provider Telco Indonesia", fontsize=18, y=1.0)
plt.xlabel('Tanggal')
plt.ylabel('Rupiah (IDR)')
plt.xticks(tanggal1,rotation='45')
plt.grid(True)
plt.legend(bbox_to_anchor=(0.,1.0,1.,.78),loc='lower center',ncol=4, mode='expand', shadow=True)
plt.show()
```

```python
# tanpa indexing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dfXl = pd.read_csv('EXCL.JK.csv',parse_dates=['Date'])
dfXl = dfXl.replace(0,np.NaN)
dfXl = dfXl.fillna(method='ffill')

dfFren = pd.read_csv('FREN.JK.csv',parse_dates=['Date'])
dfFren = dfFren.replace(0,np.NaN)
dfFren = dfFren.fillna(method='ffill')

dfIsat = pd.read_csv('ISAT.JK.csv',parse_dates=['Date'])
dfIsat = dfIsat.replace(0,np.NaN)
dfIsat = dfIsat.fillna(method='ffill')

dfTlkm = pd.read_csv('TLKM.JK.csv',parse_dates=['Date'])
dfTlkm = dfTlkm.replace(0,np.NaN)
dfTlkm = dfTlkm.fillna(method='ffill')

tanggal1 = pd.date_range(start='2019-03-31', end='2019-05-01', freq='D')
k1 = dfXl[dfXl['Date'].between('2019-04-01','2019-04-30')]
k2 = dfFren[dfFren['Date'].between('2019-04-01','2019-04-30')]
k3 = dfIsat[dfIsat['Date'].between('2019-04-01','2019-04-30')]
k4 = dfTlkm[dfTlkm['Date'].between('2019-04-01','2019-04-30')]

plt.figure(figsize=(12,5), dpi = 75)
plt.plot(k1['Date'],k1['Close'], color = 'green', label = 'PT XL AXIATA TBK' )
plt.plot(k2['Date'],k2['Close'], color = 'lightblue', label = 'PT Smartfren Telecom TBK')
plt.plot(k3['Date'],k3['Close'], color = 'blue', label = 'PT Indosat TBK')
plt.plot(k4['Date'],k4['Close'], color = 'red', label = 'PT Telekomunikasi Indonesia TBK')
plt.suptitle("Harga Historis Saham Provider Telco Indonesia (April 2019)", fontsize=18, y=1.0)

plt.xlabel('Tanggal')
plt.ylabel('Rupiah (IDR)')
plt.xticks(tanggal1,rotation='45')
plt.grid(True)
plt.legend(bbox_to_anchor=(0.,1.0,1.,.78),loc='lower center',ncol=4, mode='expand', shadow=True)
plt.show()
```
