```python
import numpy as np
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

```python
a = np.array([[1,-2,1],[3,1,-2],[7,-6,-1]])
b = np.array([6,4,10])

hasil = np.linalg.solve(a,b)
hasil
```

```python
plt.figure(figsize=(20,6))

#x - 2y + z = 6
p = plt.subplot(1,3,1,projection='3d')
x1 = np.array([6,0,0])
y1 = np.array([0,-3,0])
z1 = np.array([0,0,6])
p.scatter(x1,y1,z1,marker='o',s=100)
verts = [list(zip(x1,y1,z1))]
srf = Poly3DCollection(verts,alpha=.5,facecolor='#800000')
plt.gca().add_collection3d(srf)
plt.title('x-2y+z = 6',size=15)

#3x + y - 2z = 4
p1 = plt.subplot(1,3,2,projection='3d')
x2 = np.array([1.33,0,0])
y2 = np.array([0,4,0])
z2 = np.array([0,0,-2])
p1.scatter(x2,y2,z2,marker='o',s=100,color='red')
verts1 = [list(zip(x2,y2,z2))]
srf1 = Poly3DCollection(verts1,alpha=.5,facecolor='green')
plt.gca().add_collection3d(srf1)
plt.title('3x+y-2z = 4',size=15)

# 7x - 6y - z = 10
p2 = plt.subplot(1,3,3,projection='3d')
x3 = np.array([1.43,0,0])
y3 = np.array([0,-1.67,0])
z3 = np.array([0,0,-10])
p2.scatter(x3,y3,z3,marker='o',s=100,color='green')
verts2 = [list(zip(x3,y3,z3))]
srf2 = Poly3DCollection(verts2,alpha=.5,facecolor='blue')
plt.gca().add_collection3d(srf2)
plt.title('7x-6y-z = 10',size=15)

p.set_xlabel('Nilai X'); p1.set_xlabel('Nilai X'); p2.set_xlabel('Nilai X')
p.set_ylabel('Nilai Y'); p1.set_ylabel('Nilai Y'); p2.set_ylabel('Nilai Y')
p.set_zlabel('Nilai Z'); p1.set_zlabel('Nilai Z'); p2.set_zlabel('Nilai Z')

plt.suptitle('SPLTV',size=35)
plt.show()
```
