```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dfbcg = pd.read_csv('Balita Terimunisasi BCG 1995-2017.csv',na_values=['n.a'])
dfbcg = dfbcg.interpolate(method='linear')

dfcampak = pd.read_csv('Balita Terimunisasi Campak 1995-2017.csv',na_values=['n.a'])
dfcampak = dfcampak.interpolate(method='linear')

dfdpt = pd.read_csv('Balita Terimunisasi DPT 1995-2017.csv',na_values=['n.a'])
dfdpt = dfdpt.interpolate(method='linear')

dfpolio = pd.read_csv('Balita Terimunisasi Polio 1995-2017.csv',na_values=['n.a'])
dfpolio = dfpolio.interpolate(method='linear')

plt.figure(figsize=(11,11), dpi = 70)

plt.subplot(2,2,1)
plt.bar(dfbcg['Tahun'],dfbcg['% Balita yang pernah mendapat imunisasi BCG'],color='red')
plt.title('BCG',color='black')
plt.xticks(np.arange(1995,2018,1),rotation='90')

plt.subplot(2,2,2)
plt.bar(dfcampak['Tahun'],dfcampak['% Balita yang pernah mendapat imunisasi Campak'],color='green')
plt.title('Campak',color='black')
plt.xticks(np.arange(1995,2018,1),rotation='90')

plt.subplot(2,2,3)
plt.bar(dfdpt['Tahun'],dfdpt['% Balita yang pernah mendapat imunisasi DPT'],color='yellow')
plt.title('DPT',color='black')
plt.xticks(np.arange(1995,2018,1),rotation='90')

plt.subplot(2,2,4)
plt.bar(dfpolio['Tahun'],dfpolio['% Balita yang pernah mendapat imunisasi Polio'],color='blue')
plt.title('Polio',color='black')
plt.xticks(np.arange(1995,2018,1),rotation='90')

plt.suptitle('Persentasi balita terimunisasi 1995-2017',color='black', fontsize='12')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dfbcg = pd.read_csv('Balita Terimunisasi BCG 1995-2017.csv',na_values=['n.a'])
dfbcg = dfbcg.interpolate(method='linear')
dfbcg1 = 100 - dfbcg['% Balita yang pernah mendapat imunisasi BCG']

dfcampak = pd.read_csv('Balita Terimunisasi Campak 1995-2017.csv',na_values=['n.a'])
dfcampak = dfcampak.interpolate(method='linear')
dfcampak1 = 100 - dfcampak['% Balita yang pernah mendapat imunisasi Campak']

dfdpt = pd.read_csv('Balita Terimunisasi DPT 1995-2017.csv',na_values=['n.a'])
dfdpt = dfdpt.interpolate(method='linear')
dfdpt1 = 100 - dfdpt['% Balita yang pernah mendapat imunisasi DPT']

dfpolio = pd.read_csv('Balita Terimunisasi Polio 1995-2017.csv',na_values=['n.a'])
dfpolio = dfpolio.interpolate(method='linear')
dfpolio1 = 100 - dfpolio['% Balita yang pernah mendapat imunisasi Polio']

plt.figure(figsize=(11,11), dpi = 70)

plt.subplot(2,2,1)
plt.bar(dfbcg['Tahun'],dfbcg1,color='red')
plt.title('BCG',color='black')
plt.xticks(np.arange(1995,2018,1),rotation='90')

plt.subplot(2,2,2)
plt.bar(dfcampak['Tahun'],dfcampak1,color='green')
plt.title('Campak',color='black')
plt.xticks(np.arange(1995,2018,1),rotation='90')

plt.subplot(2,2,3)
plt.bar(dfdpt['Tahun'],dfdpt1,color='yellow')
plt.title('DPT',color='black')
plt.xticks(np.arange(1995,2018,1),rotation='90')

plt.subplot(2,2,4)
plt.bar(dfpolio['Tahun'],dfpolio1,color='blue')
plt.title('Polio',color='black')
plt.xticks(np.arange(1995,2018,1),rotation='90')

plt.suptitle('Persentasi balita tak terimunisasi 1995-2017',color='black', fontsize='12')
plt.show()
```
