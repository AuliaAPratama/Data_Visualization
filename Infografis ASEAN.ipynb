## INFOGRAFIS ASEAN

```python
import pandas as pd
import matplotlib.pyplot as plt

query = "select name, population from country where region = 'Southeast Asia' order by name"
alamatDB = "mysql://root:qiwigazper123@localhost:3306/world"
df = pd.read_sql(query, alamatDB)

plt.style.use('ggplot')
plt.figure(figsize=(12,5), dpi = 100, facecolor='w', edgecolor='b')
bebas = plt.bar(df['name'],df['population'],color=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Populasi Negara ASEAN', fontdict={'family':'impact','size':10,'color':'black'})

for i,v in enumerate(bebas):
    plt.text(v.get_x()-0.01,v.get_height()*1.01,df.iloc[i]['population'],color='brown')
plt.xlabel('Negara', fontsize=8)
plt.ylabel('Populasi (x100jt jiwa)', fontsize=8)
plt.show()
```

### PERSENTASE POPULASI ASEAN

```python
import pandas as pd
import matplotlib.pyplot as plt

query = "select name, population from country where region = 'Southeast Asia' order by name"
alamatDB = "mysql://root:qiwigazper123@localhost:3306/world"
df = pd.read_sql(query, alamatDB)

plt.style.use('ggplot')
plt.figure(figsize=(8,7), dpi = 100, facecolor='w', edgecolor='b')
bebas = plt.pie(df['population'],labels=df['name'],colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'],shadow=True,
       textprops={'size':12,'color':'black'},autopct='%1.1f%%')
plt.title('Persentase Penduduk ASEAN', fontdict={'family':'impact','size':25,'color':'black'})

plt.xlabel('Negara', fontsize=12)
plt.ylabel('Populasi (x100jt jiwa)', fontsize=12)
plt.show()
```

### GNP ASEAN

```python
import pandas as pd
import matplotlib.pyplot as plt

query = "select name, GNP from country where region = 'Southeast Asia' order by name"
alamatDB = "mysql://root:qiwigazper123@localhost:3306/world"
df = pd.read_sql(query, alamatDB)

plt.style.use('ggplot')
plt.figure(figsize=(12,5), dpi = 100, facecolor='w', edgecolor='b')
bebas = plt.bar(df['name'],df['GNP'],color=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Pendapatan Bruto Nasional ASEAN', fontdict={'family':'impact','size':10,'color':'black'})

for i,v in enumerate(bebas):
    plt.text(v.get_x()-0.03,v.get_height()*1.01,df.iloc[i]['GNP'],color='black')
plt.xlabel('Negara', fontsize=8)
plt.ylabel('Gross National Product (US$)', fontsize=8)
plt.show()
```

### Persentase Luas Daratan ASEAN

```python
import pandas as pd
import matplotlib.pyplot as plt

query = "select name,surfacearea from country where region = 'Southeast Asia' order by name"
alamatDB = "mysql://root:qiwigazper123@localhost:3306/world"
df = pd.read_sql(query, alamatDB)

plt.style.use('ggplot')
plt.figure(figsize=(8,7), dpi = 100, facecolor='w', edgecolor='b')
bebas = plt.pie(df['surfacearea'],labels=df['name'],colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'],shadow=True,
       textprops={'size':12,'color':'black'},autopct='%1.1f%%')
plt.title('Persentase Luas Daratan ASEAN', fontdict={'family':'impact','size':25,'color':'black'})

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.show()
```
